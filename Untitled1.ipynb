```python
import numpy as np
import matplotlib.pyplot as plt


# Fixing random state for reproducibility
np.random.seed(19680801)
fig, ax = plt.subplots(facecolor='black', figsize=(25, 25))
import matplotlib

plt.rcParams.update({
    "lines.color": "white",
    "patch.edgecolor": "white",
    "text.color": "black",
    "axes.facecolor": "white",
    "axes.edgecolor": "lightgray",
    "axes.labelcolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",
    "grid.color": "lightgray",
    "figure.facecolor": "black",
    "figure.edgecolor": "black",
    "savefig.facecolor": "black",
    "savefig.edgecolor": "black",
    'font.family' : 'Consolas',
    'font.weight' : 'bold',
    'font.size'   : 60,})
plt.style.use("dark_background")

# Compute pie slices
N = len(df)
theta = np.linspace(np.pi / 2, -3* ((np.pi/2)), N, endpoint=False)
radii = np.array(df['Percentage'])
width = (2 * np.pi)/ N
colors = plt.cm.viridis(radii / 10.)

ax = plt.subplot(111, projection='polar')
bars=ax.bar(theta, radii, width=width, bottom=10.0, color=colors, alpha=1.0)
for bar in bars:
    bar.set_edgecolor("black")
    bar.set_linewidth(5)
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.grid(False)
ax.axis(False)
plt.show()
```
